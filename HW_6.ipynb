#### Question 1 - Read in the data set and print the % empty values for each column. Are there any columns with missing values?

```python
data_df=pd.read_csv('C:\\Users\\Meire\\Documents\\College\\4th year\\CS301\\groundhog.csv', index_col=0)
```

```python
def get_percent_nans(df):
    for col in df.columns:  
        specific = df[col]
        whole=len(df)
        nan_len=(len(df.loc[(specific.isna())]))
        percentage= (nan_len/whole)*100
        print('The {} column has {:.2f}% empty values'.format(col,percentage))
```

```python
get_percent_nans(data_df)
```

In this table there are a couple of columns with missing values. As you can see there are the same percentage of missing in all the Average Temperature columns which happens because the first 8 years on record from 1886 to 1894 are empty for all Temperature columns.
