```python
import pandas as pd 
pd.set_option('display.max_columns', 999)
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
my_df=pd.read_csv('C:\\Users\\Meire\\Documents\\College\\4th year\\CS301\\student_class.csv')
```

```python
majors_midterm= my_df.groupby('Majors')[["Midterm"]].median().reset_index()
majors_final = my_df.groupby('Majors')[["Final"]].median().reset_index()
majors_final_avg = my_df.groupby('Majors')[["Final Average"]].median().reset_index()

label_locations = np.arange(majors_midterm.shape[0])
labs = ['BNFO','CS','IS','IT','MATH']

w = .25
plt.bar(label_locations , majors_midterm['Midterm'], width=w, label='Midterm', color='red')
plt.bar(label_locations + w, majors_final['Final'], width=w, label='Final', color='green')
plt.bar(label_locations + 2*w, majors_final_avg['Final Average'], width=w, label='Final Avg', color='blue')
plt.gcf().set_size_inches(12, 6)
plt.title('Median Midterm, Final, and Final Average Scores')
plt.xlabel('Major')
plt.ylabel('Median Score')
plt.xticks(label_locations+w, labels=labs)
plt.legend(loc='upper right')
plt.show()
```

Data items: Median grades of Midterm, Final, and Final Avg

Attributes: color red for Midterm, green for final, and blue for final avg (Categorical), (quantitive) median scores based on major  

Marks: each bar area/line

Channels: The height and colors of the bars relativity between midterm final and final avg.

Comments: The overall performance of BNFO and IS majors is that they do well on the Midterm, they do better on the final, but in the end their average goes down in comparison to their midterm and final grades. The overall performance of CS, IT, and MATH majors is that they do well on the Final, better on the Midterm, but in the end their average goes down in comparison to their midterm and final grades. In this case i would say that overall the BNFO majors are doing the best because of their final overall grade. But if we were analyzing individually then then BNFO would also be the best in final grades and and midterm grades.  

```python
numeric_bin= my_df.groupby('Numeric Bin').count().reset_index()
d={}
for idx in numeric_bin.index:
    key=numeric_bin['Numeric Bin'][idx]
    d[key]=numeric_bin['Majors'][idx]
    #x.append(numeric_bin['Majors'][idx])
x=[]
for key in d:
    i=0
    while i<d[key]:
        x.append(key)
        i=i+1
bins = [40,50,60,70,80,90,100]
plt.title('Number of Students in Each Grade Bin')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.hist(x,bins)
```

Data items: Number of students in each grade set between certain tens place 

Attributes: (Ordinal) Size of each so called bar in the histogram orders the students grade into bins. Frequency of grades 

Marks: Area of Bars, I would add a line going through it

Channels: The height of each bar on the y-axis between each section of the x-axis

Critique: I would change the naming convention of the y-axis. I would personally name it # of Students so then the title also makes more sense also I would either change the color on each level change or make defining lines for each column. I would also include a polynomial trend line to make it easily distiguished that there is a left skewed distribution.
